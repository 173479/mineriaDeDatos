IPython Magic Commands

```python
#mide el tiempo en el que se ejecuta la instruccion y mide el desempeño en
# promedio sirve para expresiones de una sola linea al usarse con solo un %
%timeit L = [n ** 2 for n in range(1000)]
```

```python
#sirve para lo mismo que el anterior pero para expresiones de mas de una linea
# por eso se usa doble %%
%%timeit
L = []
for n in range(1000):
  L.append(n ** 2)
```

Understanding Data Types in Python

```python
#importa numpy y renombralo como np (numpy genera arrays de n dimensiones)
import numpy as np
```

```python
#arreglo lineal de enteros
np.array([1, 4, 2, 5, 3])
```

```python
#al ingresar enteros y flotantes los convierte en flotantes
np.array([3.14, 4, 2, 3])
```

```python
#genera los elementos segun el tipo que le definamos
np.array([1, 2, 3, 4], dtype='float32')
```

```python
np.array([1, 2, 3, 4], dtype='int')
```

```python
#genera un rango dando el inicio y el limite en este caso dando un caso en especifico, generando una matriz bidimencional
np.array([range(i, i + 3) for i in [2, 4, 6]])
```
